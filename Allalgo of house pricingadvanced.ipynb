# Model Evaluation

```python
x = df.drop('SalePrice',axis=1)
y = df['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=120)
```

```python
x.dtypes
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
lir=LinearRegression()
```

```python
lir.fit(x_train,y_train)
```

```python
y_pred=lir.predict(x_test)
```

```python
r2_score(y_pred,y_test)
```

# SVR

```python
from sklearn.svm import SVR
svr=SVR(kernel='linear')
```

```python
svr.fit(x_train,y_train)
```

```python
y_pred=svr.predict(x_test)
```

```python
r2_score(y_pred,y_test)*100
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
```

```python
dt.fit(x_train, y_train)
```

```python
y_pred=dt.predict(x_test)
```

```python
r2_score(y_pred,y_test)*100
```

# ANN

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import MinMaxScaler
ann=Sequential()
nm=MinMaxScaler()
```

```python
x_train=nm.fit_transform(x_train)
x_test=nm.fit_transform(x_test)
```

```python
x.shape
```

```python
ann.add(Dense(74))


ann.add(Dense(64,activation='relu'))
ann.add(Dense(32,activation='relu'))
ann.add(Dense(16,activation='relu'))
ann.add(Dense(8,activation='relu'))
ann.add(Dense(4,activation='relu'))

ann.add(Dense(1))
```

```python
# Using 'mean_squared_error' as a feedback layer for regression in deep learning
ann.compile(loss='mse')
```

```python
ann.fit(x_train,y_train,epochs=500)
```

```python
ann.evaluate(x_test,y_test)
```

```python
y_pred=ann.predict(x_test)
```

```python
r2_score(y_test,y_pred)*100
```

From the above models we can see that Linear Regression model gives the best accuracy so we will use the Linear Regression algorithim for predicting the values.
